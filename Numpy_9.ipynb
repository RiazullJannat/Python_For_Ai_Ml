**Arithmetic operator and mathematical functions**

```python

x = np.array([10, 8, 16, 40])
y = np.array([2, 3, 4, 5])

add = x+y
print(add)
sub = x-y
print(sub)
mul = x*y
print(mul)
div = x/y
print(div)
```

```python
sin_val =np.sin(x)
print(sin_val)

cos_val = np.cos(x)
print(cos_val)

#radent to degree
deg_con = np.rad2deg(x)
print(deg_con)
```

```python
#log
log_val_10 = np.log10(x)
print(log_val_10)
log_val_2 = np.log2(x)
print(log_val_2)


#squire root
squire_root = np.sqrt(x)
print(squire_root)
#
```

```python
##counting

a = np.random.randint(1,100,(100,))

cnt = np.count_nonzero(a<1)
print(cnt)

unique_value, count = np.unique(a, return_counts=True)
print(unique_value)
print(count)
```

**Statistical function**

```python
data = np.genfromtxt('./sample_data/student_scores.csv', delimiter=',', skip_header=1)
math_marks=data[::,:1]
print(math_marks.T)
print("min-> ",np.min(math_marks))
print("max-> ", np.max(math_marks))

print("average->>", np.mean(math_marks))

print("median->>", np.median(math_marks))

#standard deviation
std_math_marks=np.std(math_marks)
print("standard deviation->>", std_math_marks)
```

```python
study_hours= np.array([2,4,5,7,8])
exam_scors = np.array([65,75,78,88,92])

data = np.array([study_hours,exam_scors])
print(data)
correlation = np.corrcoef(data)
print(correlation)
```

**Liner Algebra**

```python
a = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

b = np.array([
    [7,8],
    [9,10],
    [11,12]
])

dot_product=np.dot(a,b)
print(dot_product)

print(np.trace(b))
```

```python
sq_mat = np.random.randint(1,10,(3,3))
det_of_sq= np.linalg.det(sq_mat)
print(det_of_sq)
#rank
rank_sq = np.linalg.matrix_rank(sq_mat)
print(rank_sq)
```
