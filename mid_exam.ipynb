```python
import numpy as np
import pandas as pd

np.random.seed(24)

num_employees = 50  # you can increase this if needed

# Employee data
emp_ids = np.arange(101, 101 + num_employees)
emp_ages = np.random.randint(20, 55, num_employees)
emp_income = np.random.randint(35000, 95000, num_employees)

# Fixed set of roles (repetition allowed)
roles_list = [
    "HR Manager",
    "Software Developer",
    "Accountant",
    "Marketing Executive",
    "Sales Executive",
    "Backend Developer",
    "HR Executive",
    "Financial Analyst",
    "Digital Marketer",
    "Operations Manager"
]

emp_roles = np.random.choice(roles_list, num_employees)

# Create DataFrame
data = pd.DataFrame({
    "emp_id": emp_ids,
    "emp_age": emp_ages,
    "emp_salary": emp_income,
    "role": emp_roles
})

# Save dataset
data.to_csv("staff_data.csv", index=False)
```

Q4.
Create a new NumPy array where each employee’s age is adjusted based on:

ans:

```python
avj = np.mean(emp_ages)
new_arr = np.where(emp_ages < avj, emp_ages + 4, emp_ages + 2)

print("Original emp_ages array:")
print(emp_ages)

print("\nUpdated emp_ages array (new_arr):")
print(new_arr)

print(f"\nAverage age used for condition: {avj:.2f}")
```

Using NumPy:

Ans:

```python
tech_roles = ['Software Developer', 'Backend Developer','Digital Marketer']

old_total_salary_expense = np.sum(data['emp_salary'])
print(f"Old Total Salary Expense: {old_total_salary_expense:,.2f}")

modified_data = data.copy()

modified_data['emp_salary'] = np.where(modified_data['role'].isin(tech_roles), modified_data['emp_salary'] *0.08, modified_data['emp_salary'])

new_total_salary_expense = np.sum(modified_data['emp_salary'])
print(f"New Total Salary Expense: {new_total_salary_expense:,.2}")

print(f"overall increase {new_total_salary_expense - old_total_salary_expense:,.2f}")

```

Q10.
Using .loc:

ans:

```python
roles = ['Backend Developer', 'Software Developer']

data = df_file.copy()

data["previous_salary"] = data['emp_salary']

data.loc[data['role'].isin(roles), "new_salary"] = data['emp_salary'] * 1.12

data.loc[~data['role'].isin(roles), "new_salary"] = data['emp_salary'] * 1.05

data["increase_of_salary"]=data["new_salary"]-data["previous_salary"]

data[["role","previous_salary","new_salary","increase_of_salary"]]
```
