```python
class Phone:
  category = "Electronic"
  def __init__(self, name, price, color, battery_percent=100):
    self.name = name
    self.price = price
    self.color = color
    self.battery_percent = battery_percent

  def charge(self, hour):
    self.battery_percent += hour
    print(self.battery_percent)

  def capture(self, photo):
    if(self.battery_percent<=0):
      print("no charge")
    else:
      self.battery_percent -=photo
      print("photo captured...")

apple = Phone("Iphone", 10, "red")
```

```python
class Cooling_machanism:
  def __init__(self, cooling_type):
    self.cooling_type = cooling_type

  def cool(self):
    print(f"cooling... {self.cooling_type}")
```

```python
class Smartphone(Phone):
  def __init__(self, name, price, color, battery_percent, camera):
    super().__init__(name, price, color, battery_percent)
    self.camera = camera

  def charge(self, hour):
    print("fast charging...")
    super().charge(hour)
```

```python
pro = Smartphone("samsung", 800,"blue", 90, "40mah")
print(pro.camera)
pro.capture(20)
pro.charge(30)
```

```python
class SmartPhone_Cooling(Smartphone, Cooling_machanism):
  def __init__(self, name, price, color, battery_percent, camera, cooling_type):
    Smartphone.__init__(self, name, price, color, battery_percent, camera)
    Cooling_machanism.__init__(self, cooling_type)
```

```python
proMax = SmartPhone_Cooling("samsung", 700,"blue", 60, "40", "water")
print(proMax.camera)
```
