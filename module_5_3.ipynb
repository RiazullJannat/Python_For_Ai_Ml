```python
class Phone:
  category = "Electronic"
  def __init__(self, name, price, color, battery_percent=100):
    self.name = name
    self.price = price
    self.color = color
    self.battery_percent = battery_percent

  def charge(self, hour):
    self.battery_percent += hour
  def capture(self, photo):
    if(self.battery_percent<=0):
      print("no charge")
    else:
      self.battery_percent -=photo
      print("photo captured...")

apple = Phone("Iphone", 10, "red")
print(apple.category)
print(apple.name)
print(apple.price)
print(apple.color)
print(apple.capture(100))
print(apple.capture(1))
```
