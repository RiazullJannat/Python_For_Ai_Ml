# Task 2

```python
import numpy as np

def batch_generator(data,batch_size):
  for i in range(0, len(data),batch_size):
    yield data[i:i+batch_size]
```

```python
data = np.arange(20)
loader = batch_generator(data,5)
print(next(loader))
print(next(loader))
```

# Task 4

```python
def power(base, exponent=0):
  if not exponent:
    print(base*base)
  else :
    print(base**exponent)
```

```python
power(5,3)
```
