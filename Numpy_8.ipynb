```python
import numpy as np
arr = np.array([1, 2,3])
arr=arr*5
print(arr)
```

```python
arr1 = np.array([2, 3, 4])
```

```python
arr2 = np.array([
    [1, 2,3],
    [4, 5, 6]
])
```

```python
arr3 = np.array([
    [[1,2,3],[4, 5,6]],
    [[1,2,3],[4, 5,6]],
])
```

```python
#dimentation
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#data type
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)

#shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

#size
print(arr1.size)
print(arr2.size)
print(arr3.size)
```

```python
arr = np.array([1, 2, 3,"44"])
print(arr.dtype)
```

```python
arr = np.array([1, 2, 300])
print(arr.dtype)
arr = arr.astype(np.int8)
print(arr.dtype)
print(arr)
```

```python
s = {1, 2, 3}
arr = np.array(list(s))
print(arr)
print(arr.dtype)
```

```python
dc = {"a":1, "b":2}
arr = np.array(list(dc.items()))
print(arr)
print(arr.dtype)
print(arr.ndim)
print(arr.size)
print(arr.shape)
```

```python
#generate array with random value
arr = np.random.rand(2,3)
print(arr)
```

```python
arr= np.random.randint(4,10,(2,2))
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.size)
print(arr.ndim)
```

```python
# random value with uniform
arr = np.random.uniform(1,10,(3,2,2))
print(arr)
```

```python
# array using range functions
arr = np.arange(1,10, 1).reshape(3,3)
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.size)
print(arr.ndim)
```

```python
# np.linespace(star, end, how much points we want)
arr = np.linspace(1, 10, 5)
print(arr)
print(arr.dtype)
print(arr.shape)
print(arr.size)
print(arr.ndim)
```

```python
# np.logspace(start,end, how much points we want)
arr = np.logspace(0, 4, 5,base=2)
print(arr)
print(arr.dtype)
```
